### Data Imports

The data file was located at subfolder `data`.

```python
import altair as alt
import pandas as pd
import numpy as np
import os

# alt.data_transformers.enable("data_server")

# Load data into DataFrame from local csv file
tree_raw=pd.read_csv("./data/small_unique_vancouver.csv")
```

```python
tree_raw.info()
print("\n")
tree_raw.describe()
```

```python
tree_df = tree_raw[["tree_id", "genus_name"]]
tree_df
```

```python
top3_genus_list=tree_df.groupby(["genus_name"])["tree_id"].aggregate(["count"]).sort_values(by="count", ascending=False).head(3).reset_index()["genus_name"].to_list()
top3_genus_list.sort()
```

```python
top3_genus_list
```

```python
top_genus_bar=alt.Chart(tree_df[tree_df["genus_name"].isin(top3_genus_list)]).mark_bar().encode(
    alt.Y("genus_name", title="Genus"),
    alt.X("count()", title="Count of Trees")
).properties(
    title="Fig1. Trees Distribution by Top 3 Genuses"
)
top_genus_bar
```

We can see from the above visualization that there is a fairly clear the ACER is the genus which was planted the most in Vacouver street. Cherry trees are classified into genus PRUNUS. From the figure 1, the Prunus is the secend rank.
